```python
import matplotlib as plt
from matplotlib import pyplot
% matplotlib inline
import numpy as np
import seaborn as sns
import pandas as pd
filename = 'C:/Users/19177/Documents/IBM Applied Data Science/test.xlsx'
orig = pd.read_excel(filename)
df = pd.DataFrame(orig)
df
```

```python
df.dropna(subset =['# of Engagements by Head of State'], axis = 0, inplace = True)
df
```

```python
sns.regplot(x = 'Total Budget for the Royal House', y = 'Total Engagements by all Royals', data = df)
pyplot.title('Correlation between Total Royal Budget and Engagement \n')
from scipy import stats
x = df['Total Budget for the Royal House']
y = df['Total Engagements by all Royals']
cor_coef, p_val = stats.pearsonr(x, y)
print "\n The correlation coefficient is " + str(cor_coef) + "and the p-value is " + str(p_val)
print '\n The bounds of total engagements are: ' + str(pyplot.ylim(0,))
```
